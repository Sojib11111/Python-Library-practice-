```python
df['Size']=df['Size'].replace('Varies with device',np.nan)
```

```python
# First, define a function to handle 'k' and 'M' correctly
def convert_size(x):
    x = str(x).strip()
    if 'M' in x:
        return float(x.replace('M','')) * 1_000_000
    elif 'k' in x or 'K' in x:
        return float(x.replace('k','').replace('K','')) * 1_000
    else:
        return float(x)

# Apply to the column
df['Size'] = df['Size'].apply(convert_size)
```

```python
character_to_remove = ['+',',','$']
clean=['Installs','Price']
for i in character_to_remove:
    for j in clean:
        df[j] = df[j].str.replace(i,'')
```

```python
df['Size']=df['Size'].replace('Varies with device',np.nan)
```

```python
# First, define a function to handle 'k' and 'M' correctly
def convert_size(x):
    x = str(x).strip()
    if 'M' in x:
        return float(x.replace('M','')) * 1_000_000
    elif 'k' in x or 'K' in x:
        return float(x.replace('k','').replace('K','')) * 1_000
    else:
        return float(x)

# Apply to the column
df['Size'] = df['Size'].apply(convert_size)
```

```python
neumeric_features = [feature for feature in df.columns if df[feature].dtype != 'object']
categorical_features = [feature for feature in df.columns if df[feature].dtype == 'object']
print("Numeric Features:", neumeric_features)
print("Categorical Features:", categorical_features)
```

```python
df.info()
```

```python
for col in categorical_features:
    print(df[col].value_counts(normalize=True)*100)
    print('............................')
```

```python
plt.figure(figsize=(15,10))
plt.suptitle('Univariate Analysis of Numeric Features', fontsize=16,fontweight='bold',alpha=0.8,y=1.0)

for i in range(len(neumeric_features)):
    plt.subplot(3,3,i+1)
    sns.kdeplot(x=df[neumeric_features[i]], shade=True,color='r')
    plt.xlabel(neumeric_features[i])
    plt.ylabel('Density')
    plt.tight_layout()
plt.show()
   
```

```python
plt.figure(figsize=(15,10))
plt.suptitle('Univariate Analysis of Categorical Features', fontsize=16,fontweight='bold',alpha=0.8,y=1.0)
category=['Type','Content Rating']
for i in range(len(category)):
    plt.subplot(2,2,i+1)
    sns.countplot(x=df[category[i]], palette='Set2')
    plt.xlabel(category[i]) 
    plt.xticks(rotation=45)
    plt.ylabel('Count')
    plt.tight_layout()
plt.show()
```

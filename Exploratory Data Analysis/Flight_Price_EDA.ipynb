```python
df["Date_of_Journey"] = pd.to_datetime(df["Date_of_Journey"], format="%d/%m/%Y")
```

```python
df["Arrival_Time"]=df["Arrival_Time"].apply(lambda x:x.split(' ')[0])
```

```python
df
```

```python

from cmath import nan

df["Total_Stops"]=df[["Total_Stops"]].replace({"non-stop":0,"1 stop":1,"2 stops":2,"3 stops":3,"4 stops":4, nan:1})

    
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop("Route",axis=1,inplace=True)
```

```python
df
```

```python
df['d_hour'] = df['Duration'].apply(lambda x: x.split(' ')[0] if ' ' in x else x)
df['d_min'] = df['Duration'].apply(lambda x: x.split(' ')[1] if ' ' in x else 0)
```

```python
df['d_hour']=df['d_hour'].apply(lambda x: int(x.split('h')[0]) if 'h' in x else 0)

```

```python
df['d_min']=df['d_min'].apply(lambda x: int(x.split('m')[0]) if 'm' in str(x) else int(x) if str(x).isdigit() else 0)
```

```python
df.info()
```

```python
df
```

```python
df['Duration']=df['d_hour']*60+df['d_min']
```

```python
df.info()
```

```python
df
```

```python
df.rename(columns={'Duration': 'Duration(min)'}, inplace=True)
```

```python
df
```

```python
df.drop(columns=["d_hour", "d_min"], axis=1, inplace=True)
```

```python
df
```

```python
df.info()
```

```python
df["Total_Stops"]=df["Total_Stops"].astype(int)
```

```python
df['Dep_Time'] = pd.to_datetime(df['Dep_Time']).dt.strftime('%H:%M')
df['Arrival_Time'] = pd.to_datetime(df['Arrival_Time']).dt.strftime('%H:%M')
```

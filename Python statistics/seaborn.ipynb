```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='day', style='time')
plt.title('Scatter Plot of Total Bill vs Tip')
plt.show()

```

```python
sns.lineplot(x='size', y='total_bill', data=tips, marker='o', color='orange')
plt.title('Line Plot of Total Bill vs Size')
plt.show()
```

```python
sns.barplot(x="day",y="total_bill",data=tips)
plt.title('Bar Plot of Total Bill by Day')
plt.show()
```

```python
tips["day"].unique()
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Box Plot of Total Bill by Day')
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=tips,color='pink')
plt.title('Violin Plot of Total Bill by Day')
plt.show()
```

```python
sns.histplot(tips['total_bill'], bins=20, kde=True, color='green')
plt.title('Histogram of Total Bill')    
plt.show()
```

```python
sns.kdeplot(tips['total_bill'], shade=True, color='red')
plt.title('KDE Plot of Total Bill')
plt.show()
```

```python
sns.pairplot(tips, hue='day')
```

```python
corr=tips[["total_bill","tip","size"]].corr()
corr

```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

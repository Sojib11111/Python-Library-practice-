```python
import numpy as np
arr=np.array([1, 2, 3, 4, 5])
arr=list(map(lambda x:x**2,arr))
print(np.array(arr))
arrr=arrr.reshape(5,)
print(arrr)
ar=arr+arrr
print(np.array(ar))
arr=np.concatenate((arr,arrr))
print(arr)
loc=(np.where(arr==25))
print(np.array(loc))
```

```python
#Normalization
data=np.array([1, 2, 3, 4, 5])

mean=np.mean(data)
print(mean)
std=np.std(data)
print(std)
normalized_data=(data-mean)/std
print(normalized_data)
median=np.median(data)
print(median)
variance=np.var(data)
print(variance)

```

```python
data=np.array([1, 2, 3, 4, 5])
for i in range(len(data)):
    if data[i]>3 and data[i]<5:
        print(data[i])
```

```python
data=np.array([1, 2, 3, 4, 5])
arr=data[(data>3) & (data<5)]
print(arr)
```

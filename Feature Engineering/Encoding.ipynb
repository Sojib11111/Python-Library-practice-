### OneHotEncoding


```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df=pd.DataFrame({
    "color": ["red", "blue", "green", "red", "blue"]
})
```

```python
df.head()
```

```python
encoder=OneHotEncoder()
```

```python
encoded=encoder.fit_transform(df[["color"]]).toarray()
```

```python
import pandas as pd
encoded=pd.DataFrame(encoded, columns=encoder.get_feature_names_out(["color"]))
```

```python
encoded
```

```python
pd.concat([df, encoded], axis=1)
```

### Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df=pd.DataFrame({
    "size": ["S", "M", "L", "XL", "S"]
})
```

```python
df
```

```python
encoder=OrdinalEncoder(categories=[["S", "M", "L", "XL"]])
encoded=encoder.fit_transform(df[["size"]])
encoded
```

```python
encoded.tranfrom([["M"]])
```

### Target Guided Ordinal Encoding

```python
import pandas as pd
df=pd.DataFrame({
    "city": ["New York", "Los Angeles", "Chicago", "Houston", "Phoenix","New York", "Los Angeles", "Chicago", "Houston", "Phoenix"],
    "price": [100, 150, 120, 130, 110, 105, 155, 125, 135, 115]
})
```

```python
df
```

```python
mean_price=df.groupby("city")["price"].mean().to_dict()
```

```python
mean_price
```

```python
df["city_encoded"]=df["city"].map(mean_price)
```

```python
df
```

```python
df[['price', 'city_encoded']]
```

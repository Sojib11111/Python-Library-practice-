```python
import numpy as np
```

```python
lst_marks=[45,32,78,80,32,89,90,54,58,99,69,74,83,67,45,32,78,80,32,89,90,54,58,99,66,74,83,67]
min,Q1,Q2,Q3,max=np.quantile(lst_marks,[0,0.25,0.50,0.75,1.0])
print(min,Q1,Q2,Q3,max)
 
```

```python
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
print(lower_bound, upper_bound)
```
